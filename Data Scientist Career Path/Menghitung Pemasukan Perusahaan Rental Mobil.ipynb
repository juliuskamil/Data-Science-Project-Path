**Bisnis Rental Mobil**

Memiliki 6 Mobil (Avanza, Xenia, Rush, Xpander, Fortuner, Pajero)

Karena peraturan Ganjil-Genap mobil yang bisa digunakan hanya plat nomor yang berakhiran sesuai peraturan Ganjil-Genap.

Maka pemilik rental mobil, membuat list sesuai plat nomor ganjil-genap nya.
Mobil plat Ganjil = Avanza[3211], Xpander[1241], Fortuner[2345]
Mobil plat Genap = Xenia[6246], Rush[2442], Pajero[1438]

Dengan Harga Rental per hari nya Rp 800.000

Berapakah pemasukan yang diterima selama sebulan dengan pemakaian seluruh mobil sesuai peraturan ganjil-genap? 

```python
# Data
uang_jalan = 800000
jumlah_hari = 31
list_plat_nomor = [3211, 1241, 2345, 6246, 2442, 1438]
```

```python
# Pengecekan kendaraan dengan nomor pelat ganjil atau genap
kendaraan_genap = 0
kendaraan_ganjil = 0
for plat_nomor in list_plat_nomor:
    if plat_nomor % 2 == 0:
      kendaraan_genap += 1
    else:
      kendaraan_ganjil += 1
```

```python
# Total pengeluaran untuk kendaraan dengan nomor pelat ganjil dan genap dalam 1 bulan
i = 1
total_pengeluaran = 0
while i <= jumlah_hari:
  if i % 2 == 0:
      total_pengeluaran += (kendaraan_genap * uang_jalan)
  else:
      total_pengeluaran += (kendaraan_ganjil * uang_jalan)
  i += 1
```

```python
# Pengeluaran selama sebulan 
print(total_pengeluaran)
```
